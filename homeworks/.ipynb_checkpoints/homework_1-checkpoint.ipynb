```python
def get_language(page):
    res = re.search('[a-z][a-z].wikipedia.org', page)
    if res:
        return res[0][0:2]
    return 'na'
```

```python
df_train.rename(columns={"Visits": "pred_Visits"}, inplace=True)
```

```python
df_test = pd.concat([df_test, df_train["pred_Visits"]], axis=1)
```

```python
df_test.head()
```

```python
def pandas_smape(df):
    df.fillna(0, inplace=True)
    df["SMAPE"] = 200 * np.abs(df["Visits"] - df["pred_Visits"]) / (df["Visits"] + df["pred_Visits"])
    df["SMAPE"].fillna(0, inplace=True)
    return np.mean(df["SMAPE"])
```

```python
round(pandas_smape(df_test), 2)
```

```python
new_train['date'] = pd.to_datetime(new_train['date'])
new_test['date'] = pd.to_datetime(new_test['date'])
```

```python
new_train['Month']     = new_train["date"].dt.month
new_train['Day']       = new_train["date"].dt.day
new_train['DayOfWeek'] = new_train["date"].dt.dayofweek

new_test['Month']     = new_test["date"].dt.month
new_test['Day']       = new_test["date"].dt.day
new_test['DayOfWeek'] = new_test["date"].dt.dayofweek
```

```python
new_train["weekend"] = ((new_train["date"].dt.dayofweek) // 5 == 1).astype(int)
new_test["weekend"] = ((new_test["date"].dt.dayofweek) // 5 == 1).astype(int)
```

```python
median_weekend_baseline = new_train.loc[new_train["date"] > '2016-08-01'].groupby(['Page', 'weekend']).median().reset_index()
median_weekend_baseline.rename(columns={"Visits": "pred_Visits"}, inplace=True)
```

```python
pandas_smape(new_test.merge(median_weekend_baseline, on=["Page", "weekend"], how='left'))
```

```python
from dateutil.relativedelta import relativedelta
```

```python
windows = [7, 10, 12, 14, 25, 30, 60, 120, 240, 360, 365]
```

```python
for w in windows:
    start_day = new_train['date'].max() - relativedelta(days=w)
    median_weekend_baseline = new_train.loc[new_train["date"] > start_day].groupby(['Page', 'weekend']).median().reset_index()
    median_weekend_baseline.rename(columns={"Visits": "pred_Visits"}, inplace=True)
    print(w, pandas_smape(new_test.merge(median_weekend_baseline, on=["Page", "weekend"], how='left')))
```
